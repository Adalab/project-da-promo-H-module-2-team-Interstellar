```python
def llamar_pelis(url, headers, querystring):
    response = requests.get(url, headers=headers, params=querystring)
    return response.json()
```

```python
llamar_pelis(url, headers, querystring)
```

```python
# Obtiene datos de la primera página para extraer el total de páginas
json_pelis = llamar_pelis(url, headers, querystring)
total_paginas = json_pelis['page'] 
```

```python
resultados_drama = [] # Establecemos la lista donde seguardarán los datos filtrados

# Función para limpiar los datos
def limpiar_json(json_pelis):
    lista_drama = []
    if 'results' in json_pelis:
        for peli in json_pelis['results']:
            
            nombre_peli = peli.get('titleText', {}).get('text', 'Unknown')
            id_peli = peli.get('id', 'Unknown')
            lanza_peli = peli.get('releaseDate', {})
            año_peli = lanza_peli.get('year', 'Unknown')
            mes_peli = lanza_peli.get('month', 'Unknown')
            lista_drama.append(("Película", nombre_peli, año_peli, mes_peli, id_peli))
    return lista_drama
```

```python
# Iterar sobre todas las páginas de resultados
for pagina in range(1,10):
    querystring = querystring.copy()
    querystring['page'] = pagina
    json_pelis = llamar_pelis(url, headers, querystring)
    resultados_drama.extend(limpiar_json(json_pelis))
```

```python
# Mostrar resultados
print(len(resultados_drama))
print(resultados_drama)
```
